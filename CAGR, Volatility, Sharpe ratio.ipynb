```python
import numpy as np
import pandas as pd
```

```python
dat=pd.read_csv('nifty50.csv')
```

```python
dat
```

```python
dat.drop(['SBILIFE.NS','HDFCLIFE.NS'],axis=1,inplace=True)
```

```python
dat.dropna(inplace=True)
```

```python
dat
```

```python
dat.set_index('Date',inplace=True)
dat
```

```python
name = list(dat)
name
```

```python
df=pd.DataFrame()
```

```python
df['Name']=name
```

```python
def volat(ser):
    l=ser.pct_change()
    return l.std()*np.sqrt(252)
```

```python
df['Volatility']=df['Name'].apply(lambda x:volat(dat[x]))
df
```

```python
def CAGR(data):
    init=data[0]
    fin=data[-1]
    timep=5
    val = (fin/init)**(1/timep)
    return (val-1)*100
```

```python
df['CAGR']=df['Name'].apply(lambda x:CAGR(dat[x]))
df
```

```python
df['Sharpe Ratio']=((df['CAGR']-6)/df['Volatility'])
df
```
